# Masking

```python
beach = cv2.imread('beach.jpg')
cv2.imshow('beach',beach)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
mask_pic = np.zeros(beach.shape[:2],dtype='uint8')
cv2.imshow('mask',mask_pic)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
cx,cy = mask_pic.shape[1]//2, mask_pic.shape[0]//2
cv = cv2.rectangle(mask_pic,(cx-200,cy-200),(cx+200,cy+200),255,-1)    # making center white
c=cv2.bitwise_not(cv)
cv2.imshow('rect',c)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
masking = cv2.bitwise_not(beach)
cv2.imshow('masking',beach)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
ccc= cv2.bitwise_and(beach,beach, mask=c)
cv2.imshow('ccc',ccc)
cv2.waitKey()
cv2.destroyAllWindows()
```
